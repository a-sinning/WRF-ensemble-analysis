```python
#Code used to append WRF model metadata (for a given grid) to ensemble mean files (which have the same grid)
#Metadata is needed for some plotting applications
```

```python
import netCDF4 as nc
import os

dom = 2 #set the domain that we want to take metadata from and append to respective ensemble mean files... different domains have different metadata!
dom = str(dom)

# Directory where ensemble mean files are stored
ensemble_mean_directory = '/scratch/amanda/pgw_ivan_ensembles/fut_ensmean_ivan/'
# Path to original WRF file from which to copy global attributes/metadata
original_file = '/scratch/amanda/pgw_ivan_ensembles/new_pgw_ens1_nudged/wrfout_d0'+dom+'_2004-09-16_00:00:00'

# Function to copy global attributes from one file to another
def copy_global_attributes(src_file, dst_file):
    with nc.Dataset(src_file) as src, nc.Dataset(dst_file, 'r+') as dst:
        for attr in src.ncattrs():
            dst.setncattr(attr, src.getncattr(attr))
        print(f"Global attributes copied to {dst_file}")

# Read global attributes from the original file
with nc.Dataset(original_file) as src:
    global_attrs = {attr: src.getncattr(attr) for attr in src.ncattrs()}

# Loop over each ensemble mean file in the directory and copy the global attributes
for filename in os.listdir(ensemble_mean_directory):
    if filename.startswith("future_ensmean_wrfout_d0"+dom+"_") and filename.endswith(".nc"):  # Adjust condition as needed
        ensemble_file_path = os.path.join(ensemble_mean_directory, filename)
        copy_global_attributes(original_file, ensemble_file_path)
        
```

```python
#check that the global attributes/metadata were appended

# Open the NetCDF file
file_path = '/scratch/amanda/pgw_ivan_ensembles/fut_ensmean_ivan/future_ensmean_wrfout_d02_2004-09-18_18:00:00.nc' 
ds = nc.Dataset(file_path)

# Print out global attributes
print("Global attributes in the file:")
for attr in ds.ncattrs():
    print(f"{attr}: {ds.getncattr(attr)}")

# Close dataset
ds.close()
```
